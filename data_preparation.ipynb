```python
# 
import pandas as pd
import numpy as np

df = pd.read_excel('data/Base Datos Talento Tech Bogotá_Enviada.xlsx')
df.head()
```

```python
df.shape
```

# Eliminar columnas

```python
# Eliminar columnas específicas
columnas_a_eliminar = [
    'id',
    'Codigo_departamento',
    'Departamento',
    'Región',
    'Codigo_municipio',
    'Municipio'
]

df_cleaning = df.drop(columns=columnas_a_eliminar)

print("Compromiso_10_horas:" , df_cleaning['Compromiso_10_horas'].value_counts())

# Se elimina la variable Compromiso_10_horas porque todos los valores son SI
df_cleaning = df_cleaning.drop(columns=['Compromiso_10_horas'])

# Verificar resultado
df_cleaning.head()

# Acepta_requisitos_convocatoria
print("Acepta_requisitos_convocatoria:" , df_cleaning['Acepta_requisitos_convocatoria'].value_counts())

# Se elimina la variable Acepta_requisitos_convocatoria porque todos los valores son SI
df_cleaning = df_cleaning.drop(columns=['Acepta_requisitos_convocatoria'])

print("Autoriza_manejo_datos_personales:" , df_cleaning['Autoriza_manejo_datos_personales'].value_counts())

# Se elimina la variable Autoriza_manejo_datos_personales porque todos los valores son SI
df_cleaning = df_cleaning.drop(columns=['Autoriza_manejo_datos_personales'])

# Variable Disponibilidad_Equipo cambiar NO por 0 y SI por 1
df_cleaning['Disponibilidad_Equipo'] = df_cleaning['Disponibilidad_Equipo'].map({'NO': 0, 'SI': 1})

print("Presento prueba inicio:" , df_cleaning['Presento prueba inicio'].value_counts())

# Se elimina la variable Presento prueba inicio porque todos los valores son SI
df_cleaning = df_cleaning.drop(columns=['Presento prueba inicio'])

# Origen (MINTIC, CYMETRIA)
print("Origen (MINTIC, CYMETRIA):" , df_cleaning['Origen (MINTIC, CYMETRIA)'].value_counts())

# Se elimina la variable Origen (MINTIC, CYMETRIA) porque todos los valores son MINTIC
df_cleaning = df_cleaning.drop(columns=['Origen (MINTIC, CYMETRIA)'])
```

```python
df_cleaning.shape
```

```python
# mostrar que datos hay en la columna Eje_tematico
df_cleaning['Eje_tematico'].value_counts()

```

# eliminar observaciones

```python
print("Matriculado (SI o NO):" , df_cleaning['Matriculado (SI o NO)'].value_counts())

# Definir las columnas que deben estar vacías
cols_vacias = [
    'Estado (FORMADO, NO APROBADO, INACTIVO, EN FORMACION)',
    'Programa de Formación (Nombre del Curso)',
    'Cohorte (1,2,3,4,5,6,7 o 8)',
    'Tipo de formación (PRESENCIAL,VIRTUAL,HIBRIDO)',
    'Nivel',
    'Asistencia horas',
    'Total_horas posibles'
]

# Filtrar filas donde Matriculado == 'NO'
filtro_no_matriculado = df_cleaning['Matriculado (SI o NO)'].str.strip().str.upper() == 'NO'

# Verificar vacíos (NaN o celdas vacías) en todas las columnas especificadas
filtro_vacias = df_cleaning[cols_vacias].isna() | (df_cleaning[cols_vacias].astype(str).apply(lambda x: x.str.strip()) == '')

# Filas donde todas esas columnas están vacías
filtro_todas_vacias = filtro_vacias.all(axis=1)

# Combinar ambas condiciones
filtro_final = filtro_no_matriculado & filtro_todas_vacias

# Contar cuántas cumplen
cantidad = filtro_final.sum()
print(f"Número de filas con 'Matriculado = NO' y todas las columnas vacías: {cantidad}")

# Todas las filas que tienen no en la variable Matriculado (SI o NO), están vacías las últimas columnas
```

```python
# eliminar todos los no matriculados
df_cleaning = df_cleaning[df_cleaning['Matriculado (SI o NO)'] == "SI"]
```

```python
# Opción 1: quitar todas las que estén null en el eje temático
# Quitar columna Eje_tematico filas que digan Null
df_cleaning = df_cleaning[df_cleaning['Eje_tematico'] != 'Null']
df_cleaning = df_cleaning.dropna(subset=['Eje_tematico'])
df_cleaning.shape
```

```python
# En columna Estrato cambiar valores No estratificado por 0
df_cleaning['Estrato'] = df_cleaning['Estrato'].replace('No estratificado', 0)

print("NIvel_educacion:" , df_cleaning['NIvel_educacion'].value_counts())

# En variable NIvel_educacion cambiar Educación Primaria por 1, Educación Media por 2, Educación Secundaria Básica por 3,
# Educación Técnica Profesional por 4, Educación Tecnológica por 5, Educación Universitaria Pregrado por 6, 
# Especialización por 7, y quitar los que no reportan (No reporta)
nivel_educacion_map = {
    'Educación Primaria': 1,
    'Educación Media': 2,
    'Educación Secundaria Básica': 3,
    'Educación Técnica Profesional': 4,
    'Educación Tecnológica': 5,
    'Educación Universitaria Pregrado': 6,
    'Especialización': 7
}

df_cleaning = df_cleaning[df_cleaning['NIvel_educacion'] != 'No reporta']
df_cleaning['NIvel_educacion'] = df_cleaning['NIvel_educacion'].map(nivel_educacion_map)
print("NIvel_educacion:" , df_cleaning['NIvel_educacion'].value_counts())
df_cleaning.shape
```

## Dummies

```python
# Mantener solo las filas donde la variable genero sea hombre o mujer
# df_cleaning = df_cleaning[df_cleaning['Genero'].isin(['Hombre', 'Mujer'])]

# Cambiar en la variable genero mujeres en 0 y hombres en 1
df_cleaning['Genero'] = df_cleaning['Genero'].map({'Mujer': 0, 'Hombre': 1})

print("Genero:" , df_cleaning['Genero'].value_counts())
df_cleaning.shape

# Cambiar en variable Campesino No en 0 y Si en 1
df_cleaning['Campesino'] = df_cleaning['Campesino'].map({'NO': 0, 'SI': 1})

df_cleaning.head(10)
```

```python
# variable Tipo_formacion cambiar Virtual por 0 e Híbrida por 1
df_cleaning['Tipo_formacion'] = df_cleaning['Tipo_formacion'].map({'Virtual': 0, 'Híbrida': 1})

df_cleaning.head()

```

```python
# Variable Victima_del_conflicto cambiar NO por 0 y SI por 1
df_cleaning['Victima_del_conflicto'] = df_cleaning['Victima_del_conflicto'].map({'NO': 0, 'SI': 1})

df_cleaning.head()

```

## one hot encodign

```python
df_cleaning = pd.get_dummies(
    df_cleaning,
    columns=['Autoidentificacion_Etnica'],
    prefix='etnica',
    dtype=int      # 👈 fuerza que sean 0/1
)
```

```python
# Para columnas : linea_1_des_programacion	linea_2_des_inteligecia_artificial	linea_3_des_analisis_de_datos	linea_4_des_blockchain	linea_5_des_arquitectura_en_la_nube
#  cambiar Innovador - Avanzado por 0, Explorador - Básico Básico por 1, Integrador - Intermedio por 2
df_cleaning = pd.get_dummies(
    df_cleaning,
    columns=['linea_1_des_programacion'],
    prefix='linea_1',
    dtype=int      # 👈 fuerza que sean 0/1
)

df_cleaning = pd.get_dummies(
    df_cleaning,
    columns=['linea_2_des_inteligecia_artificial'],
    prefix='linea_2',
    dtype=int      # 👈 fuerza que sean 0/1
)

df_cleaning = pd.get_dummies(
    df_cleaning,
    columns=['linea_3_des_analisis_de_datos'],
    prefix='linea_3',
    dtype=int      # 👈 fuerza que sean 0/1
)

df_cleaning = pd.get_dummies(
    df_cleaning,
    columns=['linea_4_des_blockchain'],
    prefix='linea_4',
    dtype=int      # 👈 fuerza que sean 0/1
)

df_cleaning = pd.get_dummies(
    df_cleaning,
    columns=['linea_5_des_arquitectura_en_la_nube'],
    prefix='linea_5',
    dtype=int      # 👈 fuerza que sean 0/1
)

df_cleaning.head()
```

```python
# Para columna Eje Final cambiar Programación por 0, Análisis de Datos por 1, Inteligencia artificial por 2,
# Blockchain por 3 y Arquitectura en la nube por 4 
df_cleaning = pd.get_dummies(
    df_cleaning,
    columns=['Eje Final'],
    prefix='eje_final',
    dtype=int      # 👈 fuerza que sean 0/1
)
```

```python
# Para las columnas area_1_des_alfabetizacion_datos	area_2_des_comunicacion_y_colaboracion	area_3_des_contenidos_digitales	area_4_des_seguridad	area_5_des_solucion_de_problemas
# cambiar Innovador por 0, Explorador por 1, Integrador por 2
df_cleaning = pd.get_dummies(
    df_cleaning,
    columns=['area_1_des_alfabetizacion_datos'],
    prefix='area_1',
    dtype=int      # 👈 fuerza que sean 0/1
)

df_cleaning = pd.get_dummies(
    df_cleaning,
    columns=['area_2_des_comunicacion_y_colaboracion'],
    prefix='area_2',
    dtype=int      # 👈 fuerza que sean 0/1
)

df_cleaning = pd.get_dummies(
    df_cleaning,
    columns=['area_3_des_contenidos_digitales'],
    prefix='area_3',
    dtype=int      # 👈 fuerza que sean 0/1
)

df_cleaning = pd.get_dummies(
    df_cleaning,
    columns=['area_4_des_seguridad'],
    prefix='area_4',
    dtype=int      # 👈 fuerza que sean 0/1
)

df_cleaning = pd.get_dummies(
    df_cleaning,
    columns=['area_5_des_solucion_de_problemas'],
    prefix='area_5',
    dtype=int      # 👈 fuerza que sean 0/1
)


df_cleaning.head()


```

```python
# Para variable Eje_tematico cambiar Programación por 0, Datos por 1, Inteligencia artificial por 2,
# Ciber seguridad y Blockchain por 3, Computación y arquitectura en la nube por 4 y Blockchain por 5
df_cleaning = pd.get_dummies(
    df_cleaning,
    columns=['Eje_tematico'],
    prefix='eje_tematico',
    dtype=int      # 👈 fuerza que sean 0/1
)

df_cleaning.head()
```

```python
# Para variable Programa de Formación (Nombre del Curso) cambiar vacios por 0, Análisis de datos por 1, Arquitectura en la nube por 2, 
# Blockchain por 3, Inteligencia artificial por 4, Programación por 5

df_cleaning = pd.get_dummies(
    df_cleaning,
    columns=['Programa de Formación (Nombre del Curso)'],
    prefix='Programa_de_Formación',
    dtype=int      # 👈 fuerza que sean 0/1
)

df_cleaning.head(10)
```

```python
print("Discapacidad:" , df_cleaning['Discapacidad'].value_counts())

df_cleaning['Discapacidad'] = df_cleaning['Discapacidad'].notna().astype(int)

df_cleaning.head()
```

```python
# Para variable Nivel cambiar Básico por 1, Intermedio por 2, Avanzado por 3 y vacios por 0
df_cleaning = pd.get_dummies(
    df_cleaning,
    columns=['Nivel'],
    prefix='Nivel',
    dtype=int      # 👈 fuerza que sean 0/1
)
```

```python
print("Cohorte (1,2,3,4,5,6,7 o 8):" , df_cleaning['Cohorte (1,2,3,4,5,6,7 o 8)'].value_counts())

# Para variable Cohorte (1,2,3,4,5,6,7 o 8) cambiar 2_3 por 9, 3_5 por 10, 4_5 por 11, 5_7 por 12, 6_7 por 13, 6_8 por 14 y vacios por 0
df_cleaning = pd.get_dummies(
    df_cleaning,
    columns=['Cohorte (1,2,3,4,5,6,7 o 8)'],
    prefix='Cohorte',
    dtype=int      # 👈 fuerza que sean 0/1
)

df_cleaning.head(10)
```

```python
# Para variable Tipo de formación (PRESENCIAL,VIRTUAL,HIBRIDO) cambiar Híbrida por 1, Virtual por 2 y vacios por 0
df_cleaning = pd.get_dummies(
    df_cleaning,
    columns=['Tipo de formación (PRESENCIAL,VIRTUAL,HIBRIDO)'],
    prefix='Tipo_de_formacion',
    dtype=int      # 👈 fuerza que sean 0/1
)
```

## VARIABLE DE INTERES

```python
df_cleaning['estado_formado'] = (
    df_cleaning['Estado (FORMADO, NO APROBADO, INACTIVO, EN FORMACION)']
    .eq("FORMADO")
    .astype(int)
)

df_cleaning.drop('Estado (FORMADO, NO APROBADO, INACTIVO, EN FORMACION)', axis=1, inplace=True)

df_cleaning.head(10)
```

## fechas a date time

```python
# Convertir fechas a datetime
import pandas as pd

# Convertir la columna a tipo datetime (maneja formato día/mes/año y hora)
df_cleaning['fecha_ini'] = pd.to_datetime(df_cleaning['fecha_ini'], format='%d/%m/%Y %I:%M:%S %p', errors='coerce')

# Crear nuevas columnas
df_cleaning['dia'] = df_cleaning['fecha_ini'].dt.day
df_cleaning['mes'] = df_cleaning['fecha_ini'].dt.month
df_cleaning['year'] = df_cleaning['fecha_ini'].dt.year

# Eliminar la columna original
df_cleaning = df_cleaning.drop(columns=['fecha_ini'])

# Verificar resultado
# df_cleaning[['dia', 'mes', 'year']].head()
df_cleaning.head()
```

## Numericas

```python
# En variable Asistencia horas cambiar NaN por -1
df_cleaning['Asistencia horas'] = df_cleaning['Asistencia horas'].fillna(-1).astype(int)

# En variable Total_horas posibles cambiar 159 por 1 y vacias por 0
df_cleaning['Total_horas posibles'] = df_cleaning['Total_horas posibles'].replace(159, 1).fillna(0).astype(int)
df_cleaning.head()
```
